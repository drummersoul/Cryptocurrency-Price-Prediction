```python
def handle_outliers_iqr(dataframe, column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter out rows with values outside the bounds
    filtered_df = dataframe[(dataframe[column] >= lower_bound) & (dataframe[column] <= upper_bound)]

    return filtered_df
```

```python
columns_to_handle_outliers = ['open', 'high', 'low', 'close', 'volume', 'marketCap']
```

```python
for column in columns_to_handle_outliers:
    if column in df.columns:
        df = handle_outliers_iqr(df, column)
    else:
        print(f"Column '{column}' not found in the DataFrame.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Crypto_data_info.csv')

# Specify the columns you want to visualize
columns_to_visualize = ['open', 'high', 'low', 'close', 'volume', 'marketCap']

# Create box plots for each column
for column in columns_to_visualize:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df[column])
    plt.title(f'Box Plot for {column}')
    plt.show()
```
